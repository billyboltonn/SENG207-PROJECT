## DATA STATISTICS

```python

#data for pie chart
colours = ['orange', 'beige', 'yellow', 'green', 'blue', 'red', 'violet', 'black', 'pink', 'grey', 'brown', 'purple']
explode = (0.5, 0.5, 1, 1, 1, 0, 0, 0.5, 0, 0.5, 0, 0.5)
meanValues = [mean_FA, mean_VA, mean_CA, mean_RS, mean_C, mean_FSD, mean_TSD, mean_D, mean_PH, mean_S, mean_A, mean_Q]
meanTopics = ['FA',
              'VA',
              'CA',
              'RS',
              'C',
              'FSD',
              'TSD',
              'density',
              'pH',
              'sulphates',
              'alcohol',
              'quality']
```

```python
xtitle = 'Column names'
ytitle = 'Mean values'
legendlabels = ['fixed_acidity(FA)',
               'volatile_acidity(VA)',
               'citric_acid(CA)',
               'residual_sugar(RS)',
               'chlorides(C)',
               'free_sulfur_dioxide(FSD)',
               'total_sulfur_dioxide(TSD)',
               'density',
               'pH',
               'sulphates',
               'alcohol',
               'quality']
```

```python
labels = meanTopics
sizes = meanValues
```

```python
#piechat plot of mean values of each series
plt.pie(sizes, labels=labels, explode=explode, autopct='%1.1f%%', colors=colours)
plt.axis('equal')
plt.legend(title='Index', labels=legendlabels, loc='best',)
plt.title('Mean Values for each column in a pie chart')
plt.figure(figsize=(15, 18))
plt.show()
```
